```python
import numpy as np 
import matplotlib.pyplot as plt

E0=10**18
Ec=10**13

def rand():return(np.random.uniform(0,1))


E=[E0]
X=[0]
N=[1]
T=[]
for t in range(0,int(np.log(E0/Ec)/np.log(2))):
    X.append(t+1)
    
    s=N[t]
    for i in range(len(E)):
        x=rand()
        e=0.5*x*E[i]
        e1=0.5*(1-x)*E[i]
        if e>Ec:
            T.append(e)
        if e1>Ec:
            T.append(e1)
    prosorini=len(T)
    N.append(prosorini)
    E=T
    T=[]
    

E01=10**19
Ec1=10**13
E1=[E01]
X1=[0]
N1=[1]
T1=[]
for t in range(0,int(np.log(E01/Ec1)/np.log(2))):
    X1.append(t+1)
    
    s1=N1[t]
    for i in range(len(E1)):
        x1=rand()
        e=0.5*x1*E1[i]
        e1=0.5*(1-x1)*E1[i]
        if e>Ec1:
            T1.append(e)
        if e1>Ec1:
            T1.append(e1)
    prosorini=len(T1)
    N1.append(prosorini)
    E1=T1
    T1=[]
    
    
E02=10**20
Ec2=10**13
E2=[E02]
X2=[0]
N2=[1]
T2=[]
for t in range(0,int(np.log(E02/Ec2)/np.log(2))):
    X2.append(t+1)
    
    s2=N2[t]
    for i in range(len(E2)):
        x2=rand()
        e=0.5*x2*E2[i]
        e1=0.5*(1-x2)*E2[i]
        if e>Ec2:
            T2.append(e)
        if e1>Ec2:
            T2.append(e1)
    prosorini=len(T2)
    N2.append(prosorini)
    E2=T2
    T2=[]
```

```python
plt.plot(X,np.log(N))
plt.plot(X1,np.log(N1))
plt.plot(X2,np.log(N2))
```

```python
plt.figure(
    figsize=(10, 10))
plt.plot(X, np.log(N))
plt.plot(X1,np.log(N1))
plt.plot(X2,np.log(N2))
plt.xlabel('Radiation Lengths (t)')
plt.ylabel('Number of particles (N(t))')
plt.title('Electromagnetic Shower')
plt.grid(True)
```
